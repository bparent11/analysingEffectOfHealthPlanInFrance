```python
import statsmodels.api as sm

control_group = list(dico_gc.values())
treatment_group = (total.values())

# Create a DataFrame for control and treatment groups
df_control = pd.DataFrame({
    'year': years,
    'group': 'control',
    'outcome': control_group
})

df_treatment = pd.DataFrame({
    'year': years,
    'group': 'treatment',
    'outcome': treatment_group
})

# Combine control and treatment data
df = pd.concat([df_control, df_treatment], ignore_index=True)
# Encode categorical variables
df['group'] = df['group'].map({'control': 0, 'treatment': 1})
# Define treatment year (e.g., 2019)
treatment_year = 2019
# Create a binary column for "before" (0) and "after" (1) treatment
df['time'] = (df['year'] >= treatment_year).astype(int)

# Create interaction term: group * time
df['group_time'] = df['group'] * df['time']

# Independent variables: constant, time, group, and interaction term
X = sm.add_constant(df[['time', 'group', 'group_time']])
# Dependent variable (outcome)
y = df['outcome']

# Run Difference-in-Differences regression
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
'''
Running the Regression with Time Fixed Effects
'''
def regression(df, method, log):    
    print(method,':')
    # Creating dummy variables for time fixed effects (one for each year)
    df = pd.get_dummies(df, columns=['year'], drop_first=True, dtype=float)

    if method == 'event time':
        '''
        event time: how the treatment's effect varies across time
        '''
        # Create event time dummies
        df = pd.concat([df, pd.get_dummies(df['event_time'], prefix='event_time', dtype=float, drop_first=False)], axis=1)
        # Include event time dummies in the regression (along with control group and year dummies)
        #print(df['group_time'])
        #print(df[[col for col in df.columns if 'event_time' in col]])
        #X = sm.add_constant(df[['group'] + [col for col in df.columns if 'event_time' in col]])
        df = df.drop(columns='event_time')
        
        #df[list_col] = pd.DataFrame.dot(df[[col for col in df.columns if 'event_time' in col]], df[['group_time']])
        for col in ([col for col in df.columns if 'event_time' in col]):
            df['group_' + col] = df['group_time'] * df[col]
        
        #df = pd.concat([df, df[[col for col in df.columns if 'event_time' in col]]*df['group_time']], axis=1)
        #print(df.columns)
        X = sm.add_constant(df[['group'] + [col for col in df.columns if 'event_time' in col]])
        X = X.drop(columns=['event_time_-1', 'group_event_time_-1',
                            'group_event_time_-2',
                            'group_event_time_-3',
                            'group_event_time_-4', 
                            'group_event_time_-5',
                            'group_event_time_-5'])
    else:
        '''
        Independent variables: treatment and time fixed effect
        just comparing the pre-treatment vs. post-treatment period
        '''
        #X = df[['time'] + [col for col in df.columns if 'year_' in col]]
        X = sm.add_constant(df[['group_time'] + [col for col in df.columns if 'year_' in col]])
        
    
    
    
    
    X = X.apply(pd.to_numeric, errors='coerce')

    if log:
        y = df['log_outcome']
    else:
        y = df['outcome']
    # Adding a constant for the intercept
    X = sm.add_constant(X)
    
    #X = np.asarray(X)
    #y = np.asarray(y)
    # Run the regression
    model = sm.OLS(y, X).fit()
    
    # Summary of the regression
    print(model.summary())
    #print(X)
    #print(y)
```

The regression equation is expressed as follows:

$$
Y_{i,t} = \alpha + \beta \cdot \text{group}_i + \sum_{\substack{k=-5 \\ k \neq -1}}^{4} \alpha_k \cdot \text{event\_time\_k} + \sum_{k=0}^{4} \beta_k \cdot \text{group\_event\_time\_k} + \epsilon_{i,t}
$$

**Explanation of Terms**

**$Y_{i,t}$**: The outcome for individual $i$ (treatment=1, control=0) in year $t$.

**$\alpha$**: The intercept of the model.

**$\beta$**: The coefficient associated with the `group` variable (1 for treatment, 0 for control).

**$\alpha_k$**: Coefficients for the dummy variable `event_time_k`, where $k$ corresponds to the time relative to the treatment year (2019) and $k$ is different from $-1$ (baseline year 2018).

**$\beta_k$**: Coefficients for the interaction between the group and event time (`group_event_time_k`).

**$\epsilon_{i,t}$**: The error term for individual $i$ in year $t$.
