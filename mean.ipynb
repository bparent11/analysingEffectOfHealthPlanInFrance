```python
import statsmodels.api as sm

control_group = list(dico_gc.values())
treatment_group = (total.values())

# Create a DataFrame for control and treatment groups
df_control = pd.DataFrame({
    'year': years,
    'group': 'control',
    'outcome': control_group
})

df_treatment = pd.DataFrame({
    'year': years,
    'group': 'treatment',
    'outcome': treatment_group
})

# Combine control and treatment data
df = pd.concat([df_control, df_treatment], ignore_index=True)
# Encode categorical variables
df['group'] = df['group'].map({'control': 0, 'treatment': 1})
# Define treatment year (e.g., 2019)
treatment_year = 2019
# Create a binary column for "before" (0) and "after" (1) treatment
df['time'] = (df['year'] >= treatment_year).astype(int)

# Create interaction term: group * time
df['group_time'] = df['group'] * df['time']

# Independent variables: constant, time, group, and interaction term
X = sm.add_constant(df[['time', 'group', 'group_time']])
# Dependent variable (outcome)
y = df['outcome']

# Run Difference-in-Differences regression
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Plot outcomes over time
plt.figure(figsize=(10, 6))

# Plot the control group
subset_control = df[df['group'] == 0]
plt.plot(subset_control['year'], subset_control['outcome'], label='Control Group', marker='o', color='blue')

# Plot the actual treatment group
subset_treatment = df[df['group'] == 1]
plt.plot(subset_treatment['year'], subset_treatment['outcome'], label='Treatment Group', marker='o', color='green')

# Create a theoretical treatment group (projected without treatment effect)
# Assuming the treatment group would have followed the same trend as the control group before the treatment year
pre_treatment_control_avg = subset_control[subset_control['year'] < treatment_year]['outcome'].mean()
pre_treatment_treatment_avg = subset_treatment[subset_treatment['year'] < treatment_year]['outcome'].mean()
# Calculate the difference between control and treatment before treatment
difference_pre_treatment = pre_treatment_treatment_avg - pre_treatment_control_avg

# Now project the theoretical treatment outcome by adding this constant difference to control group outcome
df_theoretical = subset_control.copy()
df_theoretical['outcome'] = df_theoretical['outcome'] + difference_pre_treatment

# Plot the theoretical treatment group
plt.plot(df_theoretical['year'], df_theoretical['outcome'], label='Theoretical Treatment (No Treatment)', 
         linestyle='--', marker='x', color='red')

# Highlight the treatment year with a vertical line
plt.axvline(x=treatment_year, color='black', linestyle='--', label='Treatment Year')

# Add labels, title, and legend
plt.xticks(years, rotation=45)
plt.xlabel('Year')
plt.ylabel('Outcome')
plt.title('Difference in Differences: Actual vs Theoretical Treatment')
plt.legend()
plt.grid(True)
plt.show()
```
