**Le panier 100 % Santé audio**



Les aides auditives sont répertoriées en deux catégories : les aides auditives de classe I composent l’offre 100 % Santé. Les aides auditives de classe II, qui sont en dehors de l’offre 100 % Santé, sont des dispositifs médicaux à prix libres.  Les équipements proposés dans ce panier sont des équipements de qualité, tant par leur esthétique – aides auditives intra-auriculaires, appareils à écouteur déporté... que par leur performance technique et par les garanties qui sont associées : trente jours minimum d’essai de l’aide auditive avant achat ; quatre ans de garantie.

**Avant le 100%santé, classes A B C D**

Ces classes sont basées sur les performances et les fonctionnalités des appareils auditifs. Les appareils de classe A sont les plus performants et sophistiqués, tandis que ceux de classe D sont les plus basiques.

**Let's classify this conso by age**

```python

df_ages = pd.DataFrame(columns=['year', 'AGE','Classe avant', 'Classe I', 'Classe II'])
df_ages_gc = pd.DataFrame(columns=['year', 'AGE', 'outcome'])
df_all_without_100sante = pd.DataFrame(columns=['year', 'AGE', 'outcome'])


ages = [0, 20, 40, 60]
line_count = 0

for i, zip_file_path in enumerate( list_zip_folders):
    # Open the zip file
    with zipfile.ZipFile(zip_file_path, 'r') as z:
        
        with z.open(list_files[i]) as file:
            # Read the CSV file with semicolon as the separator and specify the encoding
            df100sante = pd.read_csv(file, sep=';', encoding='ISO-8859-1')  # or use 'windows-1252' if necessary

            #keep only the audition part
            audition_df100sante = df100sante[(df100sante['L_SC1'] == 'AUDIOPROTHESES ET ENTRETIEN, REPARATIONS ET ACCESSOIRES POUR PROCESSEURS') |
            (df100sante['L_SC1'] == 'AUDIOPROTHESES ET ENTRETIEN, REPARATIONS ET ACCESSOIRES POUR PROCESSEUR')]

            all_without_100sante = df100sante[~df100sante['L_SC1'].str.contains('OPTIQUE|AUDIOPROTHESES ET ENTRETIEN')]

            group_control = df100sante[(df100sante['L_SC1'] == 'DMI D ORIGINE SYNTHETIQUE')]

            
            for age in ages:
                #use classes 1 & 2
                df_classe1 = audition_df100sante[(audition_df100sante['L_CODE_LPP'].str.contains('CLASSE I,'))&
                (audition_df100sante['AGE']==age)]
                df_classe2 = audition_df100sante[audition_df100sante['L_CODE_LPP'].str.contains('CLASSE II,')&
                (audition_df100sante['AGE']==age)]
                df_avant = audition_df100sante[audition_df100sante['L_CODE_LPP'].str.contains('AUDIOPROTHESE, APPAREIL')&
                (audition_df100sante['AGE']==age)]

                df_ages.loc[line_count, 'AGE'] = age
                df_ages.loc[line_count, 'year'] = years[i]
                df_ages.loc[line_count, 'Classe I'] = df_classe1['QTE'].sum()
                df_ages.loc[line_count, 'Classe II'] = df_classe2['QTE'].sum()
                df_ages.loc[line_count, 'Classe avant'] = df_avant['QTE'].sum()

                df_ages_gc.loc[line_count, 'AGE'] = age
                df_ages_gc.loc[line_count, 'year'] = years[i]
                df_ages_gc.loc[line_count, 'outcome'] = int(group_control[group_control['AGE']==age]['QTE'].sum())

                df_all_without_100sante.loc[line_count, 'AGE'] = age
                df_all_without_100sante.loc[line_count, 'year'] = years[i]
                df_all_without_100sante.loc[line_count, 'outcome'] = int(all_without_100sante[all_without_100sante['AGE']==age]['QTE'].sum())
                
                line_count += 1
```

```python
df_ages['outcome'] = df_ages['Classe I'] + df_ages['Classe II'] + df_ages['Classe avant']
df_ages_gc
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

for i, (age, group) in enumerate(df_ages.groupby('AGE')):
    ax = axes[i]
    
    ax.plot(group['year'], group['Classe I'], marker='o', linestyle='-', color='b', label='Class I ')
    ax.plot(group['year'], group['Classe II'], marker='o', linestyle='-', color='g', label='Class II ')
    ax.plot(group['year'], group['Classe avant'], marker='o', linestyle='-', color='brown', label='Classes ABCD ')
    ax.plot(group['year'], group['outcome'], marker='o', linestyle='-', color='black', label='Total')
    ax.axvline(x=2019, color="y", label="Beginning of 100% santé")
    ax.axvline(x=2020, color="black", label="Covid")
    ax.axvline(x=2021, color="red", label="Class I 100% reimbursed")
    
    ax.set_title(f'Hearing aids for AGE {age}')
    ax.set_xlabel('Year')
    ax.set_ylabel('quantity')
    ax.legend()
    ax.grid(True)
plt.savefig("plots/hearing_aids_by_age.jpg", format='jpg', dpi=300)
plt.show()
```
