```python
list_files = [
    'OPEN_LPP_2014.CSV',
    'OPEN_LPP_2015.CSV',
    'OPEN_LPP_2016.CSV',
    'OPEN_LPP_2017.CSV',
    'OPEN_LPP_2018.CSV',
    'OPEN_LPP_2019.CSV',
    'OPEN_LPP_2020.CSV',
    'OPEN_LPP_2021.CSV',
    'OPEN_LPP_2022.CSV',
    'OPEN_LPP_2023.CSV']
list_zip_folders = [
    'Open-LPP-data/base_complete/OPEN_LPP_2014.zip',
    'Open-LPP-data/base_complete/OPEN_LPP_2015.zip',
    'Open-LPP-data/base_complete/OPEN_LPP_2016.zip',
    'Open-LPP-data/base_complete/OPEN_LPP_2017.zip',
    'Open-LPP-data/base_complete/OPEN_LPP_2018.zip',
    'Open-LPP-data/base_complete/OPEN_LPP_2019.zip',
    'Open-LPP-data/base_complete/OPEN_LPP_2020.zip',
    'Open-LPP-data/base_complete/OPEN_LPP_2021.zip',
    'Open-LPP-data/base_complete/OPEN_LPP_2022.zip',
    'Open-LPP-data/base_complete/OPEN_LPP_2023.zip']

years = [n for n in range (2014, 2024)]

dict_base = {2014:0,2015:0,2016:0,2017:0,2018:0,2019:0, 2020:0, 2021:0, 2022:0, 2023:0}
dict_montant = {2014:0,2015:0,2016:0,2017:0,2018:0,2019:0, 2020:0, 2021:0, 2022:0, 2023:0}


for i, zip_file_path in enumerate( list_zip_folders):
    # Open the zip file
    with zipfile.ZipFile(zip_file_path, 'r') as z:
        
        with z.open(list_files[i]) as file:
            # Read the CSV file with semicolon as the separator and specify the encoding
            df = pd.read_csv(file, sep=';', encoding='ISO-8859-1')  # or use 'windows-1252' if necessary
            
            filtered_df = df[(df['L_SC1'] == 'AUDIOPROTHESES ET ENTRETIEN, REPARATIONS ET ACCESSOIRES POUR PROCESSEURS') |
            (df['L_SC1'] == 'AUDIOPROTHESES ET ENTRETIEN, REPARATIONS ET ACCESSOIRES POUR PROCESSEUR')]
            
            filtered_df = filtered_df.copy()
            filtered_df.loc[:, 'REM_TOT'] = filtered_df['REM'] * filtered_df['QTE']
            dict_montant[years[i]] = filtered_df['REM_TOT'].sum()

            filtered_df.loc[:, 'BASE_TOT'] = filtered_df['BSE'] * filtered_df['QTE']
            dict_base[years[i]] = filtered_df['BASE_TOT'].sum()
```
